```python
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dense, Dropout

def build_sequential_multi_task_model(input_shape):
    """
    Build a multi-task Sequential model with shared and task-specific layers.
    Args:
        input_shape (tuple): Shape of the input images (height, width, channels).
    Returns:
        Model: Multi-task model with shared and task-specific outputs.
    """
    # Shared Sequential Backbone
    shared_model = Sequential([
        Input(shape=input_shape),
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        GlobalAveragePooling2D(),
        Dropout(0.5),
    ])

    # Define input layer explicitly
    shared_input = Input(shape=input_shape)
    shared_features = shared_model(shared_input)  # Call the Sequential model with inputs

    # Task-Specific Outputs
    handwriting_output = Dense(36, activation='softmax', name='handwriting_output')(shared_features)
    drawing_output = Dense(3, activation='softmax', name='drawing_output')(shared_features)
    coloring_output = Dense(3, activation='softmax', name='coloring_output')(shared_features)

    # Combine into a Multi-Task Model
    multi_task_model = Model(inputs=shared_input, outputs=[handwriting_output, drawing_output, coloring_output])

    return multi_task_model

# Define input shape
input_shape = (224, 224, 3)

# Build the model
multi_task_model = build_sequential_multi_task_model(input_shape)

# Compile the model
multi_task_model.compile(
    optimizer='adam',
    loss={
        'handwriting_output': 'categorical_crossentropy',
        'drawing_output': 'categorical_crossentropy',
        'coloring_output': 'categorical_crossentropy',
    },
    metrics=['accuracy']
)

# Display the model summary
multi_task_model.summary()
```

```python
import os
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Dataset directory
dataset_dir = './Augmented_Dataset'

# Data generators with augmentation
datagen = ImageDataGenerator(rescale=1.0 / 255, validation_split=0.2)

def create_custom_generators(base_dir, sub_dirs):
    """
    Create training and validation generators for custom directory structure.

    Args:
        base_dir (str): Base directory containing the dataset.
        sub_dirs (list): Subdirectories to include in the generator.

    Returns:
        tuple: Training and validation generators.
    """
    train_data = []
    train_labels = []
    val_data = []
    val_labels = []
    
    for label, sub_dir in enumerate(sub_dirs):
        full_dir = os.path.join(base_dir, sub_dir)
        if not os.path.exists(full_dir):
            print(f"Directory not found: {full_dir}")
            continue
        
        for root, _, files in os.walk(full_dir):
            for file in files:
                if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                    full_path = os.path.join(root, file)
                    if hash(file) % 5 == 0:  # Simple split: 80% train, 20% val
                        val_data.append(full_path)
                        val_labels.append(str(label))  # Convert label to string
                    else:
                        train_data.append(full_path)
                        train_labels.append(str(label))  # Convert label to string

    # Convert lists to arrays for generators
    train_df = pd.DataFrame({'filename': train_data, 'class': train_labels})
    val_df = pd.DataFrame({'filename': val_data, 'class': val_labels})
    
    train_gen = datagen.flow_from_dataframe(
        dataframe=train_df,
        x_col='filename',
        y_col='class',
        target_size=(512, 512),
        batch_size=32,
        class_mode='categorical'
    )
    val_gen = datagen.flow_from_dataframe(
        dataframe=val_df,
        x_col='filename',
        y_col='class',
        target_size=(512, 512),
        batch_size=32,
        class_mode='categorical'
    )
    return train_gen, val_gen

# Task-specific directories
drawing_dirs = [
    'ASD/Mild/3-6 years/Drawing', 'ASD/Mild/6-9 years/Drawing',
    'ASD/Moderate/3-6 years/Drawing', 'ASD/Moderate/6-9 years/Drawing',
    'Non-ASD/Normal/3-6 years/Drawing', 'Non-ASD/Normal/6-9 years/Drawing',
    'ASD with CD/Mild/3-6 years/Drawing', 'ASD with CD/Mild/6-9 years/Drawing'
]

handwriting_dirs = [
    'ASD/Mild/3-6 years/Handwriting', 'ASD/Mild/6-9 years/Handwriting',
    'ASD/Moderate/3-6 years/Handwriting', 'ASD/Moderate/6-9 years/Handwriting',
    'Non-ASD/Normal/3-6 years/Handwriting', 'Non-ASD/Normal/6-9 years/Handwriting',
    'ASD with CD/Mild/3-6 years/Handwriting', 'ASD with CD/Mild/6-9 years/Handwriting'
]

coloring_dirs = [
    'ASD/Mild/3-6 years/Coloring', 'ASD/Mild/6-9 years/Coloring',
    'ASD/Moderate/3-6 years/Coloring', 'ASD/Moderate/6-9 years/Coloring',
    'Non-ASD/Normal/3-6 years/Coloring', 'Non-ASD/Normal/6-9 years/Coloring',
    'ASD with CD/Mild/3-6 years/Coloring', 'ASD with CD/Mild/6-9 years/Coloring'
]

# Create generators for each task
train_generators = {}
val_generators = {}

train_generators['drawing'], val_generators['drawing'] = create_custom_generators(dataset_dir, drawing_dirs)
train_generators['handwriting'], val_generators['handwriting'] = create_custom_generators(dataset_dir, handwriting_dirs)
train_generators['coloring'], val_generators['coloring'] = create_custom_generators(dataset_dir, coloring_dirs)

print("Data generators created successfully.")
```

```python
def multi_task_generator(generators):
    """
    Combines multiple generators into a single generator for multi-task training.

    Args:
        generators (dict): Dictionary of generators for different tasks.

    Yields:
        Tuple: A tuple of (inputs, outputs), where inputs are the image data
               and outputs are a dictionary of task-specific labels.
    """
    while True:
        inputs = None
        outputs = {}

        for task, generator in generators.items():
            x, y = next(generator)
            if inputs is None:
                inputs = x  # Set inputs from the first generator
            outputs[task] = y  # Set task-specific labels

        yield inputs, outputs
```

```python
import os
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# Dataset directory
dataset_dir = './Augmented_Dataset'

# Step 1: Data Augmentation and Preprocessing
datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    validation_split=0.2,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    brightness_range=[0.8, 1.2],
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

def create_generators(base_dir, sub_dirs, batch_size=32):
    """
    Create training and validation generators for custom directory structure.

    Args:
        base_dir (str): Base directory containing the dataset.
        sub_dirs (list): Subdirectories to include in the generator.
        batch_size (int): Batch size for data loading.

    Returns:
        tuple: Training and validation generators.
    """
    train_data = []
    train_labels = []
    val_data = []
    val_labels = []

    for label, sub_dir in enumerate(sub_dirs):
        full_dir = os.path.join(base_dir, sub_dir)
        if not os.path.exists(full_dir):
            print(f"Directory not found: {full_dir}")
            continue
        
        for root, _, files in os.walk(full_dir):
            for file in files:
                if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                    full_path = os.path.join(root, file)
                    if hash(file) % 5 == 0:  # 80% train, 20% validation split
                        val_data.append(full_path)
                        val_labels.append(str(label))
                    else:
                        train_data.append(full_path)
                        train_labels.append(str(label))
    
    train_df = pd.DataFrame({'filename': train_data, 'class': train_labels})
    val_df = pd.DataFrame({'filename': val_data, 'class': val_labels})
    
    train_gen = datagen.flow_from_dataframe(
        dataframe=train_df,
        x_col='filename',
        y_col='class',
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=True
    )
    val_gen = datagen.flow_from_dataframe(
        dataframe=val_df,
        x_col='filename',
        y_col='class',
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )
    return train_gen, val_gen

# Task-specific directories
drawing_dirs = ['ASD/Mild/3-6 years/Drawing', 'ASD/Mild/6-9 years/Drawing']
handwriting_dirs = ['ASD/Mild/3-6 years/Handwriting', 'ASD/Mild/6-9 years/Handwriting']
coloring_dirs = ['ASD/Mild/3-6 years/Coloring', 'ASD/Mild/6-9 years/Coloring']

# Create data generators for each task
train_drawing, val_drawing = create_generators(dataset_dir, drawing_dirs)
train_handwriting, val_handwriting = create_generators(dataset_dir, handwriting_dirs)
train_coloring, val_coloring = create_generators(dataset_dir, coloring_dirs)

# Step 2: Combine Generators for Multi-Task Training
def combined_generator(gen1, gen2, gen3):
    """
    Combines three generators into one, ensuring batch alignment.
    """
    while True:
        batch1 = next(gen1)
        batch2 = next(gen2)
        batch3 = next(gen3)
        
        # Align batch sizes
        min_batch_size = min(len(batch1[0]), len(batch2[0]), len(batch3[0]))
        yield (
            batch1[0][:min_batch_size],  # Inputs
            {
                'handwriting_output': batch1[1][:min_batch_size],
                'drawing_output': batch2[1][:min_batch_size],
                'coloring_output': batch3[1][:min_batch_size]
            }
        )

train_generator = combined_generator(train_handwriting, train_drawing, train_coloring)
val_generator = combined_generator(val_handwriting, val_drawing, val_coloring)

steps_per_epoch = min(len(train_handwriting), len(train_drawing), len(train_coloring))
validation_steps = min(len(val_handwriting), len(val_drawing), len(val_coloring))

# Step 3: Define Multi-Task Model
def build_sequential_multi_task_model(input_shape):
    """
    Build a multi-task Sequential model with shared and task-specific layers.
    Args:
        input_shape (tuple): Shape of the input images (height, width, channels).
    Returns:
        Model: Multi-task model with shared and task-specific outputs.
    """
    # Shared Sequential Backbone
    shared_model = Sequential([
        Input(shape=input_shape),
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        GlobalAveragePooling2D(),
        Dropout(0.5),
    ])

    # Define input layer explicitly
    shared_input = Input(shape=input_shape)
    shared_features = shared_model(shared_input)  # Call the Sequential model with inputs

    # Task-Specific Outputs
    handwriting_output = Dense(len(handwriting_dirs), activation='softmax', name='handwriting_output')(shared_features)
    drawing_output = Dense(len(drawing_dirs), activation='softmax', name='drawing_output')(shared_features)
    coloring_output = Dense(len(coloring_dirs), activation='softmax', name='coloring_output')(shared_features)

    # Combine into a Multi-Task Model
    multi_task_model = Model(inputs=shared_input, outputs=[handwriting_output, drawing_output, coloring_output])

    return multi_task_model

input_shape = (224, 224, 3)
multi_task_model = build_sequential_multi_task_model(input_shape)

# Step 4: Compile Model
multi_task_model.compile(
    optimizer='adam',
    loss={
        'handwriting_output': 'categorical_crossentropy',
        'drawing_output': 'categorical_crossentropy',
        'coloring_output': 'categorical_crossentropy',
    },
    loss_weights={
        'handwriting_output': 1.0,
        'drawing_output': 1.0,
        'coloring_output': 1.2
    },
    metrics={
        'handwriting_output': ['accuracy'],
        'drawing_output': ['accuracy'],
        'coloring_output': ['accuracy'],
    }
)

# Step 5: Define Callbacks
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)

# Step 6: Train the Model
history = multi_task_model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=25,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    callbacks=[checkpoint, early_stopping, reduce_lr]
)

# Step 7: Save the Model
multi_task_model.save('final_model.keras')

# Plot training history
plt.figure(figsize=(12, 8))
for key in history.history:
    plt.plot(history.history[key], label=key)
plt.legend()
plt.title('Training History')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.show()
```
